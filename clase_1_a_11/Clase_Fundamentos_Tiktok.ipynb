 Ejercicio 3: Conversor de monedas
- Pide un monto en d贸lares y el tipo de cambio a soles. Si el monto es mayor a 1000,   
aplica un bono de 5% al total. Muestra el total en soles, redondeado a 2 decimales.

```python
dolares = float(input('Ingrese los dolares a cambiar: '))
tipo_cambio = 3.65

# Condicion
if dolares > 1000:
    bono = 0.05 * dolares
    # Formula de conversion
    soles = (dolares + bono) * tipo_cambio

soles = dolares * tipo_cambio
# Imprimir
print(f'De {dolares}$ = S/{soles}')
```

 Ejercicio 6: C谩lculo de descuento
- Pide el precio y cantidad de un producto. Si el total es mayor a 500,  
 aplica 10% de descuento. Muestra el monto final a pagar.

```python
# Solicitud de entrada
precio = float(input('Ingrese su precio: '))
cantidad = int(input('Ingrese la cantidad del producto: '))

# Formula
total = precio * cantidad

# Condicional
if (total > 500):
    descuento = total * 0.1
    total = total - descuento
    print('Se aplic贸 un 10% de descuento')
    print(f'El total con descuento: {total}')

else:
    print('No hay descuento')
    print(f'El total es {total}')
```

10. Crear una funci贸n que reciba una lista de strings  
 y devuelva cu谩ntos contienen cualquier letra

```python
# Soluci贸n 10
def contar_letra(lista_palabras, letra):
    contador = 0
    for palabra in lista_palabras:
        # letra.lower() == palabra[0].lower()
        if palabra.startswith(letra.lower()):
            contador += 1

    return contador
```
