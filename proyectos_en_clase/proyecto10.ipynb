# **Proyecto: Tendencias Globales en el Mercado Laboral de la Inteligencia Artificial 2025**

Este proyecto tiene como objetivo analizar las tendencias actuales en el mercado laboral global de Inteligencia Artificial (IA) para el año 2025. El dataset contiene 15,000 ofertas laborales relacionadas con puestos en IA y proporciona información detallada sobre salarios, habilidades requeridas, experiencia, localización de la empresa, nivel de trabajo remoto, tamaño de la empresa, sector industrial, educación requerida, entre otros aspectos relevantes.

Gracias a la riqueza de variables como salary_usd, experience_level, employment_type, company_location, remote_ratio, required_skills y industry, este análisis permitirá extraer insights clave sobre:

- Las profesiones mejor remuneradas en IA.

- La demanda de habilidades técnicas.

- La relación entre experiencia y salario.

- El impacto del trabajo remoto en la remuneración.

- Las regiones más activas en la contratación de talento en IA.

- El perfil ideal de un candidato en este sector.

Este análisis puede ser útil para profesionales que buscan orientar su carrera en IA, reclutadores, empresas tecnológicas y estudiantes que desean conocer las demandas actuales del mercado.

| Columna                  | Descripción en español                                                                |
| ------------------------ | ------------------------------------------------------------------------------------- |
| `job_id`                 | ID único asignado a cada oferta laboral relacionada con IA.                           |
| `job_title`              | Título del puesto ofrecido (ej. "AI Research Scientist").                             |
| `salary_usd`             | Salario anual en dólares estadounidenses (USD).                                       |
| `salary_currency`        | Moneda original en la que fue ofertado el salario.                                    |
| `experience_level`       | Nivel de experiencia requerido (`EN`: Entry, `MI`: Mid, `SE`: Senior, `EX`: Expert).  |
| `employment_type`        | Tipo de contrato (`FT`: Full-time, `PT`: Part-time, `CT`: Contract, `FL`: Freelance). |
| `company_location`       | País donde se encuentra la empresa contratante.                                       |
| `company_size`           | Tamaño de la empresa (`S`: Small, `M`: Medium, `L`: Large).                           |
| `employee_residence`     | País de residencia del trabajador o postulante.                                       |
| `remote_ratio`           | Porcentaje de trabajo remoto permitido (0 = presencial, 100 = remoto total).          |
| `required_skills`        | Conjunto de habilidades técnicas requeridas para el puesto.                           |
| `education_required`     | Nivel educativo requerido (ej. Bachelor, Master, PhD, Associate).                     |
| `years_experience`       | Años de experiencia requeridos para aplicar al puesto.                                |
| `industry`               | Industria o sector económico al que pertenece la empresa.                             |
| `posting_date`           | Fecha de publicación de la oferta laboral.                                            |
| `application_deadline`   | Fecha límite para enviar la postulación.                                              |
| `job_description_length` | Cantidad de caracteres del texto descriptivo del trabajo.                             |
| `benefits_score`         | Puntuación del atractivo del paquete de beneficios laborales (escala subjetiva).      |
| `company_name`           | Nombre de la empresa que ofrece el puesto.                                            |
