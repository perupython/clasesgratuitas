# **Visualizaci√≥n de Datos con matplotlib**

üß† ¬øQu√© es Matplotlib?

- Matplotlib es la biblioteca base de visualizaci√≥n de datos en Python.

- Permite crear desde gr√°ficos simples (l√≠neas, barras, dispersi√≥n) hasta visualizaciones avanzadas, totalmente personalizables.

- Es la ‚Äúnavaja suiza‚Äù para graficar en entornos como Jupyter, Google Colab o scripts de Python.

üß† ¬øPara qu√© sirve?

- Generar gr√°ficos de l√≠neas, barras, histogramas, diagramas de dispersi√≥n, gr√°ficos 3D y m√°s.

- Personalizar t√≠tulos, colores, estilos, ejes y leyendas al detalle.

- Guardar gr√°ficos en formatos como PNG, PDF o SVG para informes o presentaciones.

- Servir de base a otras librer√≠as (Seaborn, Pandas plot, etc.) que se apoyan en su motor de renderizado.

En pocas palabras: Matplotlib es el lienzo profesional para pintar tus datos en Python.

```python
fig = plt.figure(figsize=(6, 4), dpi=100)
ax = fig.add_subplot(1, 1, 1)
```

```python
ax = fig.add_subplot(2, 2, 1)
ax.set_title('Subplot 1')
ax = fig.add_subplot(2, 2, 2)
ax.set_title('Subplot 2')
ax = fig.add_subplot(2, 2, 3)
ax.set_title('Subplot 3')
ax = fig.add_subplot(2, 2, 4)
ax.set_title('Subplot 4')
```

```python
# iteraciones -> bucle for
fig = plt.figure(figsize=(8, 6), dpi=100)

for i in range(1, 5): # 1,2,3,4
    ax = fig.add_subplot(2, 2, i)
    ax.set_title(f'Subplot {i}')

fig.tight_layout(pad=1)
```

Primeros Pasos

```python
import numpy as np # Librer√≠a para calculo numerico
```

```python
x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
x = np.linspace(0, 2 * np.pi)
seno = np.sin(x)
coseno = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, seno, label='$f_1(x) = sin(x)$')
ax.plot(x, coseno, label='$f_2(x) = cos(x)$')
ax.legend()
ax.set_xlim(3, 5)
ax.set_ylim(-1, -0.5)
```
