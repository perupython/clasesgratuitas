```python
# Pandas: Manipulación de datos
import pandas as pd
```

```python
# Lectura de datos
df = pd.read_excel('Catalogo1960_2023.xlsx')
```

```python
# Mostrar las 5 primeras filas
df.head()
```

```python
# Selecciona una columna 'MAGNITUD'
df['MAGNITUD']
```

```python
df['MAGNITUD'].values
```

```python
# Promedio de la columna MAGNITUD
df['MAGNITUD'].mean()
```

```python
# Suma de la columna MAGNITUD
df['MAGNITUD'].sum()
```

```python
# Valor máximo de la columna MAGNITUD
df['MAGNITUD'].max()
```

```python
# Valor mínimo de la columna MAGNITUD
df['MAGNITUD'].min()
```

```python
# Matplotlib
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(df['LATITUD'], df['LONGITUD'], alpha=.2)
plt.show()
```

```python
plt.hist(df['MAGNITUD'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma de Magnitud')
plt.xlabel('Magnitud')
plt.ylabel('Frecuencia')
plt.show()
```

Área de un triángulo
- Pide base y altura, calcula el área del triángulo.

```python
# Solución
base = float(input('Base: '))
altura = float(input('Altura: '))

formula = (base * altura) / 2
print(f'El área del triángulo es {formula} mtros')
```
