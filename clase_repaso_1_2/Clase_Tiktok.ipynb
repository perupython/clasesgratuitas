2. Calculadora de propina en restaurante  
📌 Tema: Variables, tipos de datos

- El usuario ingresa el total de su cuenta en un restaurante y el porcentaje de propina que desea dejar. Calcula el total a pagar.

```python
# total de la cuenta
total_cuenta = float(input('Ingrese el total de tu cuenta: S/ '))
# porcentaje de propina
porcentaje_propina = float(input('¿Que porcentaje de propinas deseas dejar? % '))
# calcular el total a pagar
monto_pagar = total_cuenta + (total_cuenta * porcentaje_propina / 100)

# Imprimir
print(f'Total a pagar con propina: S/ {round(monto_pagar, 2)}')
```

3. Listado de productos comprados en el mercado  
📌 Tema: Listas

- Solicita al usuario que ingrese 5 productos comprados. Luego imprime la lista y la cantidad total de productos.

```python
# Lista para mis productos
productos = []

# ¿Como llamar 5 veces a mi producto?
for i in range(5):
    # el valor que el usuario ingresará
    producto = input('Ingrese el producto: ')
    # Metodo de las listas para agregar elementos a tu lista
    productos.append(producto)

print(f'Productos comprados: {productos}')
print(f'Total de productos: {len(productos)}')
```
