ðŸŽ¯ Primera dinÃ¡mica: Clasificar frutas ðŸŽðŸŒ
- Imagina que queremos que la computadora aprenda a diferenciar **manzanas** y **bananas**.

```python
# Datos ficticios de frutas
frutas = [
    {"color": "rojo", "longitud": 7, "nombre": "manzana"},
    {"color": "amarillo", "longitud": 12, "nombre": "banana"},
    {"color": "rojo", "longitud": 8, "nombre": "manzana"},
    {"color": "amarillo", "longitud": 11, "nombre": "banana"},
]
```

```python
for f in frutas:
    print(f'Color: {f["color"]}, Longitud: {f["longitud"]} -> {f["nombre"]}')
```

ðŸ‘‰ La idea es que el modelo *aprenda reglas* por sÃ­ mismo:
- Si es rojo y corto â†’ Manzana  
- Si es amarillo y largo â†’ Banana  
Eso es ML: **detectar patrones automÃ¡ticamente.**

ðŸ”¹ Tipos de Machine Learning
1. **Supervisado** â†’ El modelo aprende con ejemplos etiquetados.  
   - Ej: Clasificar correos como "spam" o "no spam".  
2. **No supervisado** â†’ El modelo encuentra patrones sin etiquetas.  
   - Ej: Agrupar clientes segÃºn su comportamiento de compra.  
3. **Reforzamiento** â†’ El modelo aprende con prueba y error.  
   - Ej: Un robot que aprende a caminar.

ðŸŽ¯ Segunda dinÃ¡mica: PredicciÃ³n simple de precios
- Vamos a hacer un ejemplo con **Python + Scikit-Learn** para predecir precios de casas de forma bÃ¡sica.

```python
# Importamos librerÃ­as
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Datos ficticios: tamaÃ±o de la casa (m2) y precio (en miles de $)
X = np.array([50, 60, 80, 100, 120, 150]).reshape(-1,1) # TamaÃ±o de la casa
y = np.array([100, 120, 140, 180, 200, 240]) # Precio de la casa
```

ðŸ‘‰ En palabras: Precio = punto de inicio + (pendiente Ã— tamaÃ±o de la casa)

ðŸ”¹ Ejemplo numÃ©rico

Supongamos que queremos predecir el precio de una casa en miles de dÃ³lares:

ð‘¦ = 20 + 1.5 â‹… ð‘¥
- Intercepto (b) = 20 â†’ El precio base (cuando la casa tiene tamaÃ±o 0, valor simbÃ³lico).

- Pendiente (a) = 1.5 â†’ Cada mÂ² adicional aumenta el precio en 1.5 mil dÃ³lares.  

Si una casa mide 100 mÂ²:

- y = 20 +1.5 â‹… 100 = 20 + 150 =170Â (milÂ dÃ³lares)


```python
# Creamos el modelo
modelo = LinearRegression()
# Entrenamiento
modelo.fit(X, y)
```

```python
# PredicciÃ³n
nueva_casa = np.array([[90]])
y_pred = modelo.predict(nueva_casa)

print(f'Precio estimado para una casa de 90 m2: {y_pred}')
```

```python
# VisualizaciÃ³n
# Grafica de los datos reales
plt.scatter(X, y, color = 'blue', label='Datos Reales')
# Punto que acabamos de predecir
plt.scatter(90, y_pred, color = 'green', s=100, label='PredicciÃ³n 90 m2')
# Grafica del modelo predictivo
plt.plot(X, modelo.predict(X), color = 'red', label='Modelo')
# Titulo de etiquetas
plt.xlabel('TamaÃ±o de la casa (m2)')
plt.ylabel('Precio (miles $)')
plt.legend()
plt.show()
```

ðŸ‘‰ Este es un **modelo supervisado**. Le damos ejemplos (datos histÃ³ricos) y predice valores nuevos.
