# **¿Qué es Machine Learning y por qué está en todos lados?**

🔹 ¿Qué es Machine Learning?
- Machine Learning es una rama de la Inteligencia Artificial que permite que las computadoras **aprendan de los datos** sin estar programadas explícitamente paso a paso.

En lugar de darles reglas fijas, les damos **ejemplos** y la máquina aprende a detectar patrones.

👉 Definición sencilla: *Es enseñar a una máquina a mejorar sus resultados con la experiencia (datos).*


🔹 Ejemplos de Machine Learning en tu vida diaria
1. **Netflix** – recomienda películas y series que probablemente te gusten.  
2. **Spotify** – sugiere música parecida a lo que escuchas.  
3. **Google Maps** – calcula rutas según tráfico en tiempo real.  
4. **Correo electrónico** – filtra automáticamente el *spam*.  
5. **Bancos** – detectan fraudes revisando patrones sospechosos.  

.

🎯 Primera dinámica: Clasificar frutas 🍎🍌
- Imagina que queremos que la computadora aprenda a diferenciar **manzanas** y **bananas**.

```python
# Datos ficticios de frutas
frutas = [
    {"color": "rojo", "longitud": 7, "nombre": "manzana"},
    {"color": "amarillo", "longitud": 12, "nombre": "banana"},
    {"color": "rojo", "longitud": 8, "nombre": "manzana"},
    {"color": "amarillo", "longitud": 11, "nombre": "banana"},
]
```

```python
for f in frutas:
    print(f'Color: {f["color"]}, Longitud: {f["longitud"]} -> {f["nombre"]}')
```

👉 La idea es que el modelo *aprenda reglas* por sí mismo:
- Si es rojo y corto → Manzana  
- Si es amarillo y largo → Banana  
Eso es ML: **detectar patrones automáticamente.**

🔹 Tipos de Machine Learning
1. **Supervisado** → El modelo aprende con ejemplos etiquetados.  
   - Ej: Clasificar correos como "spam" o "no spam".  
2. **No supervisado** → El modelo encuentra patrones sin etiquetas.  
   - Ej: Agrupar clientes según su comportamiento de compra.  
3. **Reforzamiento** → El modelo aprende con prueba y error.  
   - Ej: Un robot que aprende a caminar.

.

🎯 Segunda dinámica: Predicción simple de precios
- Vamos a hacer un ejemplo con **Python + Scikit-Learn** para predecir precios de casas de forma básica.

```python
# Importamos librerías
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Datos ficticios: tamaño de la casa (m2) y precio (en miles de $)
X = np.array([50, 60, 80, 100, 120, 150]).reshape(-1,1) # Tamaño de la casa
y = np.array([100, 120, 140, 180, 200, 240]) # Precio de la casa
```

👉 En palabras: Precio = punto de inicio + (pendiente × tamaño de la casa)

🔹 Ejemplo numérico

Supongamos que queremos predecir el precio de una casa en miles de dólares:

𝑦 = 20 + 1.5 ⋅ 𝑥
- Intercepto (b) = 20 → El precio base (cuando la casa tiene tamaño 0, valor simbólico).

- Pendiente (a) = 1.5 → Cada m² adicional aumenta el precio en 1.5 mil dólares.  

Si una casa mide 100 m²:

- y = 20 +1.5 ⋅ 100 = 20 + 150 =170 (mil dólares)


```python
# Creamos el modelo
modelo = LinearRegression()
# Entrenamiento
modelo.fit(X, y)
```

```python
# Predicción
nueva_casa = np.array([[90]])
y_pred = modelo.predict(nueva_casa)

print(f'Precio estimado para una casa de 90 m2: {y_pred}')
```

```python
# Visualización
# Grafica de los datos reales
plt.scatter(X, y, color = 'blue', label='Datos Reales')
# Punto que acabamos de predecir
plt.scatter(90, y_pred, color = 'green', s=100, label='Predicción 90 m2')
# Grafica del modelo predictivo
plt.plot(X, modelo.predict(X), color = 'red', label='Modelo')
# Titulo de etiquetas
plt.xlabel('Tamaño de la casa (m2)')
plt.ylabel('Precio (miles $)')
plt.legend()
plt.show()
```

👉 Este es un **modelo supervisado**. Le damos ejemplos (datos históricos) y predice valores nuevos.

🔹 Resumen  
✅ Machine Learning significa **enseñar a las máquinas con datos**.  
✅ Lo usamos todos los días (Netflix, Spotify, Google Maps).  
✅ Hay distintos tipos: supervisado, no supervisado y reforzamiento.  
✅ Con Python podemos construir modelos predictivos en minutos.
