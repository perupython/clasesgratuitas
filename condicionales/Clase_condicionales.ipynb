ðŸ“Œ 4. CONDICIONALES CON else

```python
# Asignar una variable 'temperatura'
temperatura = 26
# Condicional: si temperatura es mayor a 25, imprime hace mucho calor y si no, imprime el clima es tolerable
if (temperatura > 25): # 26 > 25 -> True
    # Caso True
    print('Hace mucho calor')

# Caso False
else:
    print('El clima es tolerable')
```

```python
# Asignar una variable 'temperatura'
temperatura = 20
# Condicional: si temperatura es mayor a 25, imprime hace mucho calor y si no, imprime el clima es tolerable
if (temperatura > 25): # 20 > 25 -> False
    # Caso True
    print('Hace mucho calor')

# Caso False
else:
    print('El clima es tolerable')
```

ðŸ“Œ 5. CONDICIONALES CON elif

```python
# Asignar una variable 'nota'
nota = 19

# Condicional: 
# si la nota es mayor o igual que 18, imprime 'excelente'
if (nota >= 18):
    print('Excelente')
# y si nota es mayor o igual que 14, imprime 'aprobado'
elif (nota >= 14):
    print('Aprobado')
# y si no, imprime 'desaprobado'
else:
    print('Desaprobado')

print('Fin de la condicional')
```

ðŸ“Œ 6. USOS REALES DE CONDICIONALES

âœ… Ejemplo 1: Clasificar edades

```python
# Asignar una variable 'edad'
edad = int(input('Ingrese su edad: '))

# Condicional:
# Si edad es menor o igual que 12, imprime niÃ±o
if (edad <= 12):
    print('NiÃ±o')
# y si edad es menor o igual que 17, imprime adolescente
elif (edad <= 17):
    print('Adolescente')
# y si edad es menor o igual que 59, imprime adultos
elif (edad <= 59):
    print('Adulto')
# y sino, imprime adulto mayor
else:
    print('Adulto mayor')
```

ðŸ“Œ 7. CONDICIONALES ANIDADAS

```python
# asignar variables 'usuario' y 'clave'
usuario = 'admin'
clave = '12345'

# Condicional para el usuario
if usuario == 'admin':
    # Condicional para la clave
    if clave == '1234':
        print('Acceso permitido')
    else:
        print('Clave incorrecta')

else:
    print('Usuario no encontrado')
```

ðŸ“Œ 8. OPERADOR TERNARIO (condicional en una lÃ­nea)

```python
# Asignar una variable 'edad'
edad = 20
# Condicional con operador ternario
# Sintaxis: bloque True if condicion else bloque false
mensaje = 'Mayor de edad' if edad >= 18 else 'Menor de edad'
print(mensaje)
```

```python
# Validar si hay stock
stock = 0
mensaje = 'Disponible' if stock > 0 else 'Agotado'
print(mensaje)
```

```python
# 3. Clasificador de IMC (Ãndice de Masa Corporal)
peso = float(input('Ingrese tu peso: '))
altura = float(input('Ingrese tu altura: '))

# Formula
imc = peso / (altura ** 2)

if imc < 18.5:
    print('Bajo peso')

elif imc < 25:
    print('Peso normal')

elif imc < 30:
    print('Sobrepeso')

else:
    print('Obesidad')
```

```python
# 5. Clasificador de productos por categorÃ­a y precio
categoria = input('CategorÃ­a (A, B, C): ').upper()
precio = float(input('Precio del producto: '))

if categoria == 'A':
    if precio > 1000:
        print('Producto Premiun')
    else:
        print('Producto EstÃ¡ndar')

elif categoria == 'B':
    if precio > 500:
        print('Producto Semi-premiun')
    
    else:
        print('Producto Basico')

else:
    print('Categoria no vÃ¡lida o sin clasificacion')
```
