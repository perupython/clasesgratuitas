```python
# importar librerías
import seaborn as sns
import pandas as pd # manipulación de datos
import matplotlib.pyplot as plt # gráficas
```

```python
# Data ejemplo(viene con seaborn)
df = sns.load_dataset('tips')
```

```python
# Mostrar df
df.head()
```

```python
df.info()
```

```python
df['total_bill'] # Seleccionamos una columna
```

Distribución

```python
# Histograma
sns.histplot(df['total_bill'], bins=20, kde=True)
plt.show()
```

Countplot

```python
# conteo de categorías
sns.countplot(x='day', data=df)
plt.show()
```

```python
# conteo de sexo
sns.countplot(x='sex', data=df)
plt.show()
```

### **Comparaciones y relaciones**

Relación entre dos variables numéricas

```python
# Gráfico de dispersión
sns.scatterplot(data=df, x='total_bill', y='tip')
# Personalización
plt.title('Relación entre el total de la cuenta vs propina')
plt.xlabel('Total de la cuenta')
plt.ylabel('Propina')
plt.show()
```

```python
# Gráfico de dispersión
sns.scatterplot(data=df, x='total_bill', y='tip', hue='day')
# Personalización
plt.title('Relación entre el total de la cuenta vs propina')
plt.xlabel('Total de la cuenta')
plt.ylabel('Propina')
plt.show()
```

```python
# Gráfico de dispersión
sns.scatterplot(data=df, x='total_bill', y='tip', hue='smoker')
# Personalización
plt.title('Relación entre el total de la cuenta vs propina')
plt.xlabel('Total de la cuenta')
plt.ylabel('Propina')
plt.show()
```

```python
# Gráfico de dispersión
sns.scatterplot(data=df, x='total_bill', y='tip', hue='size')
# Personalización
plt.title('Relación entre el total de la cuenta vs propina')
plt.xlabel('Total de la cuenta')
plt.ylabel('Propina')
plt.show()
```

Gráfico de cajas

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.show()
```

```python
sns.boxplot(x='sex', y='total_bill', data=df)
plt.show()
```

```python
sns.boxplot(x='smoker', y='total_bill', data=df)
plt.show()
```

Gráfico de violin

```python
sns.violinplot(x='day', y='total_bill', data=df)
plt.show()
```

Pairplot: comparaciones múltiples

```python
sns.pairplot(df, hue='sex')
plt.show()
```

Mapa de calor

```python
corr = df.corr(numeric_only=True)
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.savefig('grafico1.jpg')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.scatterplot(x='total_bill', y='tip', hue='time', data=df)
plt.savefig('grafico2.jpg')
plt.show()
```

```python
df = sns.load_dataset('penguins')
df.head()
```

.

### **Distribuciones comparativas**

- Histograma múltiple:

```python
sns.histplot(data= df, x='flipper_length_mm', hue='species')
plt.title('Distribución de Longitud de aletas por especie')
plt.show()
```

### **Comparaciones entre categorías**

Boxplot avanzado

```python
sns.boxplot(x='species', y='body_mass_g', hue='sex',palette='Set3', data = df)
plt.title('Peso Corporal por Especie y Sexo')
plt.show()
```

### **Relaciones entre variables**

Scatterplot con personalización

```python
plt.figure(figsize=(9, 6))
sns.scatterplot(
    data=df, x='bill_length_mm', y='bill_depth_mm',
    hue='species', style='sex', palette='magma', s=80, alpha=0.8
)
```

Barplot con media

```python
plt.figure(figsize=(9, 6))
sns.set_style('whitegrid')
sns.barplot(x='species', y='body_mass_g', data=df, hue='sex', palette='pastel')
plt.title('Peso Promedio por Especie(con desviacion)')
plt.show()
```

Pairplot

```python
sns.pairplot(df.dropna(), hue='island', palette='Set2')
plt.show()
```

Mapa de correlación

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='viridis', lw=1)
plt.title('Mapa de Correlación - Pinguinos', pad=20, fontsize=20)
plt.show()
```
