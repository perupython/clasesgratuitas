```python
# importar librerías
import seaborn as sns
import pandas as pd # manipulación de datos
import matplotlib.pyplot as plt # gráficas
```

```python
# Data ejemplo(viene con seaborn)
df = sns.load_dataset('tips')
```

```python
# Mostrar df
df.head()
```

```python
df.info()
```

```python
df['total_bill']
```

Distribución

```python
# Histograma
sns.histplot(df['total_bill'], bins=20, kde=True)
plt.show()
```

Countplot

```python
# conteo de categorías
sns.countplot(x='day', data=df)
plt.show()
```

```python
# conteo de sexo
sns.countplot(x='sex', data=df)
plt.show()
```

### **Comparaciones y relaciones**

Relación entre dos variables numéricas

```python
# Gráfico de dispersión
sns.scatterplot(data=df, x='total_bill', y='tip')
# Personalización
plt.title('Relación entre el total de la cuenta vs propina')
plt.xlabel('Total de la cuenta')
plt.ylabel('Propina')
plt.show()
```

```python
# Gráfico de dispersión
sns.scatterplot(data=df, x='total_bill', y='tip', hue='day')
# Personalización
plt.title('Relación entre el total de la cuenta vs propina')
plt.xlabel('Total de la cuenta')
plt.ylabel('Propina')
plt.show()
```

```python
# Gráfico de dispersión
sns.scatterplot(data=df, x='total_bill', y='tip', hue='smoker')
# Personalización
plt.title('Relación entre el total de la cuenta vs propina')
plt.xlabel('Total de la cuenta')
plt.ylabel('Propina')
plt.show()
```

```python
# Gráfico de dispersión
sns.scatterplot(data=df, x='total_bill', y='tip', hue='size')
# Personalización
plt.title('Relación entre el total de la cuenta vs propina')
plt.xlabel('Total de la cuenta')
plt.ylabel('Propina')
plt.show()
```

Gráfico de cajas

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.show()
```

```python
sns.boxplot(x='sex', y='total_bill', data=df)
plt.show()
```

```python
sns.boxplot(x='smoker', y='total_bill', data=df)
plt.show()
```

Gráfico de violin

```python
sns.violinplot(x='day', y='total_bill', data=df)
plt.show()
```

Pairplot: comparaciones múltiples

```python
sns.pairplot(df, hue='sex')
plt.show()
```

Mapa de calor

```python
corr = df.corr(numeric_only=True)
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.savefig('grafico1.jpg')
plt.show()
```

```python
sns.set_style('darkgrid')
sns.scatterplot(x='total_bill', y='tip', hue='time', data=df)
plt.savefig('grafico2.jpg')
plt.show()
```
