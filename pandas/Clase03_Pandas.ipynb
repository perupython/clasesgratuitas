```python
# Cambio en las columnas
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('?', '').str.replace(':','_').str.replace('(', '').str.replace(')','').str.replace('/', '')
```

```python
df.columns
```

```python
column_remove = ['unit_id', 'air_temp-celsius', 'whoverified', 'dateverified']
# Eliminación de columnas con más del 80% de valores nulos.
df.drop(column_remove, axis=1, inplace=True)
```

```python
# Tratamiento de datos nulos
promedio_domgl = df['dissolved_oxygen_mgl'].mean()
df['dissolved_oxygen_mgl'] = df['dissolved_oxygen_mgl'].fillna(promedio_domgl)
```

```python
# Corregir los datos nulos
num_cols = ["salinity_ppt","ph_standard_units","secchi_depth_m",
            "water_depth_m","water_temp_c","air_temp_f"]
# Iteración para eliminar los valores nulos en la lista de 'num_cols'
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.dropna(inplace=True) # Eliminación de filas con datos nulos (65)
```

```python
# Convertir fechas en Pandas
df['read_date'] = pd.to_datetime(df['read_date']) # aaaa-mm-dd
```
