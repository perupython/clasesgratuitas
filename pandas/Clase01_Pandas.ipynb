🧠 ¿Qué es Pandas?
- Pandas es una biblioteca de Python para el análisis y manipulación de datos estructurados, especialmente tablas (como Excel o CSV).

🧠 ¿Para qué sirve?
- Leer archivos CSV, Excel, JSON.

- Limpiar, filtrar, agrupar y transformar datos.

- Estadísticas descriptivas rápidas.

- Ideal para ciencia de datos y análisis empresarial.

🔹 ¿Qué es una Serie?
- Una Serie es una estructura de datos unidimensional de Pandas, muy parecida a una lista o a un array de NumPy, pero con etiquetas (índices) asociadas a cada elemento.

🔸 Características:

- Tiene una sola columna de datos.

- Cada dato tiene un índice.

- Soporta operaciones vectorizadas y estadísticas.  

📌 ¿Para qué sirve?

- Para manejar datos simples con etiquetas.

- Ideal para operaciones estadísticas rápidas, como medias, sumas o filtros.

- Útil cuando queremos trabajar con una variable individual dentro de un análisis.

🔹 ¿Qué es un DataFrame?
- Un DataFrame es una estructura bidimensional (como una tabla de Excel o SQL) compuesta por múltiples Series alineadas por su índice.

🔸 Características:

- Tiene filas y columnas.

- Cada columna es una Serie.

- Es la estructura más utilizada en análisis de datos con Pandas.

📌 ¿Para qué sirve?

- Para almacenar, visualizar y analizar datos estructurados.

- Permite hacer:

    - Filtraciones y consultas.

    - Estadísticas descriptivas.

    - Agrupaciones, ordenamientos y visualizaciones.

    - Es la base para cualquier trabajo de análisis de datos con Python.
